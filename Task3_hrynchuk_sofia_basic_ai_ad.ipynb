***Завдання 5:*** В моделях машинного навчання інснує велика проблема опрацювання безперервних змінних. Потрібно перетворити безперервні значення в категоріальні. 

Максимальний вік пасажира - ```80 років```.

Отже, давайте розділимо діапазон від ```0 до 80``` на ```5``` осередків. 


Додайте до нашого датасету нову колонку ```Age_bin```, в якій будуть значення ```0, 1, 2, 3, 4```відповідно кожній віковій групі.

Побудуйте графік, який відображає кількість виживших у кожній віковій категорії.





````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')


bins = [0, 16, 32, 48, 64, 80]
df['Age_bin'] = pd.cut(df['Age'], bins, labels=[0, 1, 2, 3, 4])
survived_by_age = df.groupby('Age_bin')['Survived'].sum()
survived_by_age.plot(kind='bar', color='pink')
plt.title('Кількість виживших у кожній віковій категорії')
plt.xlabel('Вікова категорія')
plt.ylabel('Кількість виживших')
plt.xticks(rotation=0)
plt.show()
````

Ваші висновки: По нашому графіку можна зрозуміти, що кількість виживших залежить від вікової категорії

***Завдання 6:*** Так як змінна ```fare``` також є неперервною її треба закодувати до ```порядкової шкали``` 
Для цього скоритайтесь функцією ```pandas.qcut```.
[pandas.qcut](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html)  
- необхідно перевірити, чи є вірною гіпотеза про те, що кількість виживших пасажирів для квартильних значеннь змінної ```fare``` відрізняється, та чи доцільно використовувати кодування на цю змінну(для подальшої роботи з закодованою фичею)

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')


df['Fare_bin'] = pd.qcut(df['Fare'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])


survived_by_fare = df.groupby('Fare_bin')['Survived'].sum()
print("Кількість виживших пасажирів для кожного квартилю змінної fare:")
print(survived_by_fare)
survived_by_fare.plot(kind='bar', color='lightpink')
plt.title('Кількість виживших для кожного квартилю змінної fare')
plt.xlabel('Квартиль змінної fare')
plt.ylabel('Кількість виживших')
plt.xticks(rotation=0)
plt.show()
````

Ваші висновки: За діаграмою ми можемо зробити висновок,що пасажири, які мали квитки з вищими цінами, які відповідають кращим класам, мали більші шанси на виживання.

***Завдання 7:*** Необхідно створити нові змінні ```Family_size``` та ```Alone``` та проаналізувати їх. Змінна ```Family_size``` передає суму змінних ```Parch``` та ```SibSp```. Змінна ```Family_size``` буде відібражати комбіновані дані, які необхідні для перевірки гіпотези, що ```виживання пасажирів залежіть від кількості членів їх сім'ї```. Змінна ```Alone``` позначає наявніть сім'ї.

````python
import pandas as pd
df = pd.read_csv('titanic.csv')


df['Family_size'] = df['Parch'] + df['SibSp']
df['Alone'] = df['Family_size'].apply(lambda x: 1 if x == 0 else 0)


print(df[['Family_size', 'Alone']].head())
````

Ваші висновки: По результату можу сказати, що більшість людей мали якогось співмандрівника собою

***Завдання 9:*** Перетворіть змінні ```Sex```, ```Embarked``` та ```Title``` в числові значення


````python
import pandas as pd
df = pd.read_csv('titanic.csv')

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})


embarked_encoded = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_encoded], axis=1)
df.drop('Embarked', axis=1, inplace=True)

if 'Title' not in df.columns:
    df['Title'] = pd.Series(['Mr', 'Master', 'Mrs', 'Miss', 'Other'])
df['Title'] = df['Title'].map({'Mr': 0, 'Master': 1, 'Mrs': 2, 'Miss': 3, 'Other': 4})


print(df.head())

````

Ваші висновки: Вивела основну статистичну інформацію для числових змінних, можу сказати що атрибути Cabin та Age мають багато пропущених значень, також можемо побачити що кількість виживших збильшується від вартрсті квитка

***Завдання 10.*** Побудувати графіки всіх категоріальних змінних та проаналізувати скільки кожна категорія містить даних

***Завдання 11.*** Побудувати діаграми розподілу всіх числових змінних

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')


numeric_vars = ['Age', 'SibSp', 'Parch', 'Fare']
for var in numeric_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=var, kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.show()

````
